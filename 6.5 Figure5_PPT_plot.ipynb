```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设置随机分布参数
means = [0.0, 2.0, 4.0]
sigma = 0.8
colors = ['lightgreen', 'lightcoral', 'lightyellow']
labels = ['Series 1', 'Series 2', 'Series 3']

# 横坐标范围
x = np.linspace(-2, 6, 500)

plt.figure(figsize=(8, 5))

# 绘制和填充面积
for mu, color, label in zip(means, colors, labels):
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, y, color=color.replace('light', ''), linewidth=2, label=label)
    plt.fill_between(x, y, color=color, alpha=0.7)

# 美化图表
plt.xticks([])
plt.yticks([])
plt.legend(frameon=False)
plt.title('Overlap Filled Curves with Pale Green, Red, Yellow')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define theta range with more points for smoother curves
theta = np.linspace(-5, 5, 1000)

# Prior: N(0, 1)
prior = norm.pdf(theta, loc=0, scale=1)

# Generate more data points with better spread
n_points = 300  # Increased to 300 points
x_data = np.linspace(-3, 3, n_points)
y_data = 2 + np.random.normal(0, 0.2, n_points)  # Less noise for clearer pattern

# Likelihood: assume observed y=2, noise sigma=1
y_obs = 2
likelihood = norm.pdf(y_obs, loc=theta, scale=1)

# Unnormalized posterior
unnormalized_posterior = prior * likelihood
# Normalize posterior
posterior = unnormalized_posterior / np.trapz(unnormalized_posterior, theta)

# Create figure with larger size
plt.figure(figsize=(10, 6))

# Set background color to light blue
plt.gcf().set_facecolor('#E6F3FF')  # Light blue background
plt.gca().set_facecolor('white')  # White data area

# Plot prior with distinct style
plt.fill_between(theta, prior, alpha=0.2, color='lightblue', label='_nolegend_')
plt.plot(theta, prior, label='Prior', color='blue', linestyle='--', linewidth=2)

# Plot likelihood
scaled_likelihood = likelihood / np.max(likelihood) * np.max(prior)
plt.fill_between(theta, scaled_likelihood, alpha=0.2, color='lightgreen', label='_nolegend_')
plt.plot(theta, scaled_likelihood, label='Likelihood (scaled)', color='green', linewidth=2)

# Plot posterior with distinct style
plt.fill_between(theta, posterior, alpha=0.2, color='mistyrose', label='_nolegend_')
plt.plot(theta, posterior, label='Posterior', color='red', linewidth=3)

# Plot data points with better visibility
plt.scatter(x_data, y_data/np.max(y_data)*np.max(prior), 
           color='black', alpha=0.6, s=40, label='Data Points')

# Enhance plot aesthetics
plt.xlabel(r'Parameter $\theta$', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Bayesian Inference Visualization', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper left')

# Add text annotations
plt.text(3.5, 0.35, 'Prior: N(0,1)', color='blue', fontsize=10)
plt.text(3.5, 0.3, 'Posterior', color='red', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate gamma distribution parameters and data
alpha = 2  # shape parameter
beta = 1.5  # scale parameter
x = np.linspace(0, 10, 200)
gamma_dist = stats.gamma.pdf(x, a=alpha, scale=beta)

# Create figure with larger size
plt.figure(figsize=(8, 5))

# Set background color to white
plt.gca().set_facecolor('white')
plt.gcf().set_facecolor('white')

# Plot gamma distribution
plt.plot(x, gamma_dist, 'r-', lw=2, label=f'Gamma(α={alpha}, β={beta})')
plt.fill_between(x, gamma_dist, alpha=0.2, color='mistyrose')

# Enhance plot aesthetics
plt.xlabel('x', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.title('Gamma Distribution', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)

# Add text annotation
plt.text(6, 0.2, f'α (shape) = {alpha}\nβ (scale) = {beta}', 
         bbox=dict(facecolor='white', alpha=0.8), fontsize=10)

plt.tight_layout()
plt.show()
```
