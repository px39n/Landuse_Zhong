```python
# Set the file path
file_path = r'D:\xarray\aligned2\process_OSMdata\data\solar_7.gpkg'

# Get layer information using geopandas
import geopandas as gpd
import fiona
import matplotlib.pyplot as plt

# List available layers
layers = fiona.listlayers(file_path)
print("Available layers:", layers)

# Read the polygons and points layers
polygons = gpd.read_file(file_path, layer='polygons')
points = gpd.read_file(file_path, layer='points')

# Select first polygon feature
selected_polygon = polygons.iloc[[0]]

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the selected polygon
selected_polygon.plot(ax=ax, color='red', alpha=0.5)

# Get the bounds of the selected polygon
bounds = selected_polygon.total_bounds

# Filter points within the polygon bounds
points_within = points.cx[bounds[0]:bounds[2], bounds[1]:bounds[3]]

# Plot the points
points_within.plot(ax=ax, color='blue', markersize=50)

plt.title('Selected Polygon with Points')
plt.show()
```

```python
# Read and display GHS raster information
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# File path for GHS data
ghs_path = r'D:\xarray\aligned2\process_OSMdata\data\ghs_2015.tif'

# Open the raster file and get information
with rasterio.open(ghs_path) as src:
    print("Raster Information:")
    print(f"Driver: {src.driver}")
    print(f"Width: {src.width}")
    print(f"Height: {src.height}")
    print(f"Number of bands: {src.count}")
    print(f"Coordinate Reference System: {src.crs}")
    print(f"Transform: {src.transform}")
    print(f"Resolution: {src.res}")
    print(f"Nodata value: {src.nodata}")
    
    # Read and plot the data
    # Read first band and mask NoData values
    data = src.read(1, masked=True)
    
    # Calculate statistics
    print("\nData Statistics:")
    print(f"Minimum value: {np.nanmin(data)}")
    print(f"Maximum value: {np.nanmax(data)}")
    print(f"Mean value: {np.nanmean(data):.2f}")
    print(f"Median value: {np.nanmedian(data):.2f}")
    print(f"Standard deviation: {np.nanstd(data):.2f}")
    
    # Create two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10))
    
    # Plot the raster data
    im1 = ax1.imshow(data)
    plt.colorbar(im1, ax=ax1)
    ax1.set_title('GHS 2015 Data')
    
    # Plot histogram of values
    ax2.hist(data.compressed(), bins=50)
    ax2.set_title('Distribution of Values')
    ax2.set_xlabel('Value')
    ax2.set_ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
```
