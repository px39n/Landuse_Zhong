### 2、PV efficiency intergrated land suitability 


（1）光伏板能量平衡方程
$${\color{Orange}(1-\alpha-\varepsilon) R_{\downarrow}^{sun}+L_{\downarrow}^{sky}+L_{\uparrow}^{g}-2 L^{p}-2 q_{c o n v}=0,} $$

each parameter denotes as:
$\varepsilon$ denotes the efficiency of the solar panel（太阳能电池效率）；
$\alpha$ denotes PV panel surface albedo（光伏表面反射率）；$R_{\downarrow}^{sun}$ denotes shortwave radiation from the sun(短波辐射)
$L_{\downarrow}^{sky}$ refers to the integral longwave radiation reaching the solar module（长波辐射）；

（2）天空长波辐射等式
$$L_{\downarrow}^{s k y}=1.24 \sigma\left(\frac{e_{a}}{T_{a}}\right)^{\frac{1}{7}} T_{a}^{4}$$

式中， $e_a$  表示实测水汽压（百帕）， $T_a$  为实测气温（开尔文）， $\sigma = 5.670367 \times 10^{-8} \, \text{kg} \cdot \text{s}^{-1} \cdot \text{K}^{-4}$  为斯特藩-玻尔兹曼常数。来自地面的入射长波辐射采用理想黑体模型进行表征：

（3）地面长波辐射模型
$$L_{\uparrow}^{g} = T_g^4,$$
式中$T_g^4$ 表示地表温度。光伏板的长波辐射被模拟为黑体辐射

（4）光伏板长波辐射模型
$$L^p = \sigma T_p^4,$$
其中$T_p$为光伏板温度。光伏板的对流冷却过程采用总体输送方程描述：

（5）对流冷却热通量模型
$$q_{\text{conv}} = h(T_p - T_a)$$

（6）对流换热系数
$$h = 0.036 \cdot \frac{k_{\text{air}}}{l_{\text{panel}}} \cdot \left( \frac{u \cdot l_{\text{panel}}}{\nu} \right)^{4/5} \cdot Pr^{1/3}$$

$k_{\text{air}}$ denotes 0.026 $W/mk$ 代表空气导热率，$l_{\text{panel}}$ denotes length of solar panel取值为1.5米，$u$表示风速$m/s$，$\nu$表示空气运动粘度$1.57\times10^{-5}m^2/s$

（7）光伏效率温度修正方程
$$\varepsilon = \varepsilon_{ref} \left[ 1 - A (T_p - T_{ref}) \right]$$
$\varepsilon_{ref}$参考效率13.5%，$T_{ref}$参考温度是298K，$A=0.0051/K$是光伏板效率与温度的依赖关系，

在以上方程来自于[参考文献](https://www.nature.com/articles/s41598-019-47803-3)，基于此方程，需要输入的环境变量为风速、气温、太阳辐射、水蒸气压四个变量，求解太阳能板温度进而反演得出效率。
